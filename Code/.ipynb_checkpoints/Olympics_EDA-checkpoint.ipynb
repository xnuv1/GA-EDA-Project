### Data Handling: 
__Cleaning, transforming, and combining data__

```python
# Standardizing the sex varaints into two Male and Female

olympics_df['Sex'] = olympics_df['Sex'].replace(['M', 'male'], 'Male')
olympics_df['Sex'] = olympics_df['Sex'].replace(['F', 'FeMale'], 'Female')
olympics_df['Sex'].value_counts()
```

```python
# Clean Height and Weight columns 
olympics_df['Height'] = olympics_df['Height'].astype(str).str.replace('cm','').str.strip()
olympics_df['Weight'] = olympics_df['Weight'].astype(str).str.replace('kg','').str.strip()

# Seperate the 'Games' columns into two 'Year' and 'Season'
olympics_df['Year'] = olympics_df['Games'].astype(str).str.split().str[0].astype(int)
olympics_df['Season'] = olympics_df['Games'].astype(str).str.split().str[1].astype(str)
```

```python
# Change The Height, Weight and Age types
olympics_df['Height'] = pd.to_numeric(olympics_df['Height'], errors='coerce')
olympics_df['Weight'] = pd.to_numeric(olympics_df['Weight'], errors='coerce')
olympics_df['Age'] = pd.to_numeric(olympics_df['Age'], errors='coerce')
olympics_df.dtypes
```

```python
#Seperate the olympics_df into two dataframes based on the sex. This is to get more accurate median to fill in the missing values

male_df = olympics_df[olympics_df['Sex'] == 'Male'].copy()
female_df = olympics_df[olympics_df['Sex'] == 'Female'].copy()
male_df.shape
female_df.shape
```

```python
# Impute the missing age values with the median based on the sex and the country

# Age for Male
male_age_df = male_df.groupby('NOC')['Age'].transform('median')
male_df['Age'].fillna(male_age_df, inplace=True)


# Age for Female
female_age_df = female_df.groupby('NOC')['Age'].transform('median')
female_df['Age'].fillna(female_age_df, inplace=True)
```

```python
# Impute the missing Height values with the median based on the sex and the country

# Height for Male
male_height_df = male_df.groupby('NOC')['Height'].transform('median')
male_df['Height'].fillna(male_height_df, inplace=True)


# For Female
female_height_df = female_df.groupby('NOC')['Height'].transform('median')
female_df['Height'].fillna(female_height_df, inplace=True)
```

```python
# Impute the missing Weight values with the median based on the sex and the country

# Weight for Male
male_weight_df = male_df.groupby('NOC')['Weight'].transform('median')
male_df['Weight'].fillna(male_weight_df, inplace=True)


# For Female
female_weight_df = female_df.groupby('NOC')['Weight'].transform('median')
female_df['Weight'].fillna(female_weight_df, inplace=True)
```

```python
# To check if there is nulls after imputing 
male_df[['Age', 'Height', 'Weight']].isnull().sum()
female_df[['Age', 'Height', 'Weight']].isnull().sum()
```

```python
# Drop the nulls rows (some of the countries dont have any value to get the median from)
male_df = male_df.dropna(subset=['Height', 'Weight'])
female_df = female_df.dropna(subset=['Height', 'Weight'])
```

```python
male_df[['Age', 'Height', 'Weight']].isnull().sum()
female_df[['Age', 'Height', 'Weight']].isnull().sum()
```

```python
# Recombine the cleaned male & female data back into one DataFrame
olympics_df = pd.concat([male_df, female_df], ignore_index=True)
```

```python
# Change to 'No Medals' when there in null in the Medal column 
olympics_df['Medal'] = olympics_df['Medal'].fillna('No Medals')
```

## Female Comaprision 

```python
# Compare Female Demographic between 1952 and 2016

female_compare = olympics_df[
    (olympics_df['Year'].isin([1952, 2016])) & 
    (olympics_df['Season'] == "Summer") & 
    (olympics_df['Sex'] == "Female")
]
for demo in ['Age','Height','Weight']:
    plt.figure(figsize=(8,5))
    sns.histplot(data=female_compare, x=demo, hue='Year', kde=True, bins=25, palette=tableau_colors)
    plt.title(f"Female {demo} Distribution (1952 vs 2016)")
    plt.show()
```

```python
female_summary = female_compare.groupby('Year')[['Age','Height','Weight']].median()
print(female_summary)
```

## Observation:
### Age

- Women in 1952 were mostly in their late teens/early 20s.

- By 2016, more were in their mid-to-late 20s.

### Height

- Average height went up from around 160s to 170s.

- The 2016 Olympics shows a wider range.

### Weight

- In 1952, most weighed around 5565 kg.

- In 2016, it center shifted to 6575 kg with more spread.

# Male vs Female Participation

```python
# Comparing the number of participation between 1952 and 2016

compare_years = olympics_df[(olympics_df['Year'].isin([1952,2016])) & (olympics_df['Season']=="Summer")]
sns.countplot(data=compare_years, x='Year', hue='Sex', palette=colors)
plt.title("Male vs Female Participation: 1952 vs 2016 (Summer)")
plt.tight_layout()
plt.show()
```

#### Observation
- There was a significant increase in the number of female events, which largely explains the rise in female participation

```python
# Count male vs female participants by year
participation = olympics_df.groupby(['Year','Sex']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.lineplot(data=participation, x='Year', y='count', hue='Sex', marker='o', palette=colors)
plt.title("Male vs Female Participation in the Olympics Over the Years")
plt.ylabel("Number of Athletes")
plt.xlabel("Year")
plt.show()
```

#### Observation
- This Graph shows the number of participation over the time from 1896 to 2016

```python
# Zoomed-in plot

participation_between_1980_2016 = participation[
    (participation['Year'] >= 1980) & (participation['Year'] <= 2016)
]


plt.figure(figsize=(10,6))
sns.lineplot(data=participation_between_1980_2016, x='Year', y='count', hue='Sex', marker='o', palette=colors)
plt.title("Male vs Female Participation (19802016)")
plt.ylabel("Number of Athletes"); plt.xlabel("Year")
plt.show()
```

#### Observation
- This Graph shows the number of participation from 1896 until 2016

```python
# Count male vs female participants by year
gender_participation = olympics_df.groupby(['Year','Sex']).size().reset_index(name='count')

# Pivot so Male and Female are columns
gender_ratio = gender_participation.pivot(index='Year', columns='Sex', values='count').fillna(0).reset_index()

# Calculate Female-to-Male ratio
gender_ratio['Female_to_Male_Ratio'] = gender_ratio['Female'] / gender_ratio['Male']



# Plot Female/Male ratio over time
plt.figure(figsize=(10,6))
sns.lineplot(data=gender_ratio, x='Year', y='Female_to_Male_Ratio', marker='o', color="purple")
plt.title("Female to Male Athlete Ratio Over the Years (All Olympics)")
plt.ylabel("Female / Male Ratio")
plt.xlabel("Year")
plt.show()
```

#### Observation
- This Graph shows the ratio of the males and females over the years from the 1896 to 2016
- The Graphs shows that we are approching gender banlance un upcoming games

## The Last 6 Olympics

```python
last_six_years = olympics_df[olympics_df['Year'] >= 2006]
sns.countplot(data=last_six_years, x='Year', hue='Sex', palette = tableau_colors)
plt.show()
```

#### Observation
- In Summer Olympics ( 2008 , 2012 , 2016), higher participation than in the Winter
- age gap decreases continuously

```python
sns.boxenplot(data= last_six_years , x='Year' , y= 'Age', palette=tableau_colors)
plt.title('Age distribtuion in last 6 Olympics')
```

#### Observation
- The distribution of age in the last 6 Olympics

- Summer Olympics (2008, 2012, 2016) show a wider age range with more outliers, especially in events that rely on skill and precision rather than physical intensity.

- Winter Olympics tend to have a narrower age distribution, with fewer older athletes.

- Events like Equestrian Mixed Dressage and Shooting Men's Rapid-Fire Pistol (25m) attract older participants, as these disciplines emphasize technique and experience over physical conditioning.

```python
# Group by Event and collect unique genders
gender_by_event = olympics_df.groupby('Event')['Sex'].nunique()

# Filter events that have both Male and Female participants
shared_events = gender_by_event[gender_by_event == 2]

# Count how many events are shared
num_shared_events = shared_events.count()

print(f"Number of events shared between both genders: {num_shared_events}")
```

#### Observation
- There are only 58 Events that both genders participate in

```python
# Number of Unique Events by Gender  

event_counts = last_six_years.groupby('Sex')['Event'].nunique().reset_index(name='Unique_Events')

sns.barplot(data=event_counts, x='Sex', y='Unique_Events', palette=colors)
plt.title('Number of Unique Events by Gender (Last 6 Olympics)')
plt.ylabel('Event Count')
plt.show()
```

#### Observation
- The number of Events for both genders in last 6 Olympic. The male is higher

```python
# Age Distribution of Gold Medalists in Last 6 Olympics

gold_medalists = last_six_years[last_six_years['Medal'] == 'Gold']

sns.histplot(data=gold_medalists, x='Age', bins=15, kde=False, color='goldenrod')
plt.title('Age Distribution of Gold Medalists in Last 6 Olympics')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

#### Observation
- Distribution of Players that won the Gold medal. The majoirty of them are between 15 to 35

## Winter & Summer Olympics

```python
# Remove duplicate 
unique_events = olympics_df.drop_duplicates(subset=['Event', 'Sex', 'Season'])

# Count events exist by Season and by Sex
event_counts = unique_events.groupby(['Season', 'Sex'])['Event'].count().reset_index()

# Create the barplot
sns.barplot(data=event_counts, x='Season', y='Event', hue='Sex', palette=colors)
plt.title('Number of Unique Events by Gender and Season')
plt.ylabel('Event Count')
plt.xlabel('Olympic Season')
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

#### Observation
- The Summer Olympics feature roughly five times more events than the Winter Olympics, highlighting greater diversity and global participation.

```python
# Count the number of unique participants  
participant_counts = olympics_df.groupby(['Season', 'Sex'])['ID'].nunique().reset_index()

# Create the barplot
sns.barplot(data=participant_counts, x='Season', y='ID', hue='Sex', palette=colors)
plt.title('Number of Participants by Gender and Season')
plt.ylabel('Participant Count')
plt.xlabel('Olympic Season')
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

#### Observation
- The Winter Olympics have significantly fewer participants compared to the Summer Olympics, reflecting the smaller event pool and regional specialization.
